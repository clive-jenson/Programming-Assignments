## Question 1.
----
### Write a function that stutters a word as if someone is struggling to read it. The first two letters are repeated twice with an ellipsis ... and space after each, and then the word is pronounced with a question mark ?.

**Examples**

        stutter("incredible") ➞ "in... in... incredible?";

        stutter("enthusiastic") ➞ "en... en... enthusiastic?";
    
        stutter("outstanding") ➞ "ou... ou... outstanding?";

**Hint** :- Assume all input is in lower case and at least two characters long.

```python
def stutter(st):
    st.lower()
    if len(st) >= 2:
        first_two = st[:2] + '... ' 
        output = first_two * 2 + st + '?'
    return output
```

```python
print(stutter("incredible"))


stutter("outstanding")
```

### Create a function that takes an angle in radians and returns the corresponding angle in degrees rounded to one decimal place.

**Examples**

>`radians_to_degrees(1) ➞ 57.3`

>`radians_to_degrees(20) ➞ 1145.9`

>`radians_to_degrees(50) ➞ 2864.8`

```python
import math

def radians_to_degrees(rad):
    degree = round(rad* 180/math.pi,2)
    return str(degree) + '\u00B0' 
```

```python
print(radians_to_degrees(1))

print(radians_to_degrees(50))
```

## Question 3.
---
### In this challenge, establish if a given integer num is a Curzon number. If 1 plus 2 elevated to num is exactly divisible by 1 plus 2 multiplied by num, then num is a Curzon number.

### Given a non-negative integer num, implement a function that returns True if num is a Curzon number, or False otherwise.

**Examples**

>`is_curzon(5) ➞ True`

$$2  ^{5}  + 1 = 33$$
    
   $$ 2  \times  5 + 1 = 11$$
    
$$33 \text{ is a multiple of 11}$$

>`is_curzon(10) ➞ False`

$$2  ^{10}   + 1 = 1025$$

$$2  \times  10 + 1 = 21$$

$$1025 \text{ is not a multiple of 21}$$

>`is_curzon(14) ➞ True`

$$2   ^{14}  + 1 = 16385$$

$$2  \times   14 + 1 = 29$$

$$16385 \text{ is a multiple of 29}$$

```python
def is_curzon(n):
    x = (2 ** n) + 1
    y = (2 * n) + 1 
    if x % y == 0:
        return True
    
    return False
```

```python
print(is_curzon(10))

is_curzon(14) 
```

### Given the side length x find the area of a hexagon.



**Examples**

>`area_of_hexagon(1) ➞ 2.6`

>`area_of_hexagon(2) ➞ 10.4`

>`area_of_hexagon(3) ➞ 23.4`



```python
import math
# side length = s
def area_of_hexagon(s): 
    A = round((3* math.sqrt(3) * s ** 2 )/2,2) 
    return A
```

```python
print(area_of_hexagon(1))
area_of_hexagon(3)
```

### Create a function that returns a base-2 (binary) representation of a base-10 (decimal) string number. To convert is simple:

<ul>
    <li> 2 - base 2 </li>
    <li> 10 - base 10 </li>
</ul>

`010101001(2) = 1 + 8 + 32 + 128`.

Going from right to left, the value of the most right bit is 1, now from that every bit to the left
will be x2 the value, value of an 8 bit binary numbers are 
`(256, 128, 64, 32, 16, 8, 4, 2, 1)`

**Examples**

>binary(1) ➞ &quot;1&quot; 
$$1 \times 1 = 1$$

>binary(5) ➞ &quot;101&quot;
$$1\times1 + 1\times4 = 5$$

>binary(10) ➞ &quot;1010&quot;

$$ 1\times2 + 1\times8 = 10$$

```python
def binary(num):
   
    if num >= 1:
        binary(num // 2)
    print(num % 2, end = '')
```
